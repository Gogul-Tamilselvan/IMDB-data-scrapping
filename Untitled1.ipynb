```python
from bs4 import BeautifulSoup
import requests,openpyxl
```

```python
#xport data using excel
excel= openpyxl.Workbook()
sheet=excel.active
sheet.title='Movie'
sheet.append(['rank','movie_name','movie_rating','year'])

try:
    url=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
    soup = BeautifulSoup(url.text,'html.parser')
    movies = soup.find("tbody",class_="lister-list").find_all('tr')
    
    for movie in movies:
        rank = movie.find('td',class_='titleColumn').get_text().split('.')[0].strip()
        movie_name = movie.find('td',class_='titleColumn').a.text
        movie_rating = movie.find('td',class_='ratingColumn').strong.text
        year = movie.find('td',class_='titleColumn').span.text.replace('(',' ')
        #print(rank,movie_name,movie_rating,year.replace(')',' '))
        print(rank,movie_name,movie_rating,year.replace('(',' '))
        year=year.replace('(',' ')
        sheet.append([rank,movie_name,movie_rating,year])

        
except Exception as e:
    print(e)
    
excel.save('movies.xls')                                                          
```

```python
#exporting data using pandas
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
try:
    url=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250')
    soup = BeautifulSoup(url.text,'html.parser')
    movies = soup.find("tbody",class_="lister-list").find_all('tr')
    movie_list ={'Rank':[],'Name':[],'Rating':[],'Year':[]}
    for movie in movies:
        rank = movie.find('td',class_='titleColumn').get_text().split('.')[0].strip()
        movie_name = movie.find('td',class_='titleColumn').a.text
        movie_rating = movie.find('td',class_='ratingColumn').strong.text
        year = movie.find('td',class_='titleColumn').span.text.replace('(',' ')
        #print(rank,movie_name,movie_rating,year.replace(')',' '))
        year=year.replace(')',' ')
        movie_list['Rank'].append(rank)
        movie_list['Name'].append(movie_name)
        movie_list['Rating'].append(movie_rating)
        movie_list['Year'].append(year)
        
except Exception as e:
    print(e)
    
df=pd.DataFrame(movie_list)
df
    
```
